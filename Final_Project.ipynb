# **Manual Evaluation of Model Performance**

```python
import json
import pandas as pd
import spacy
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
jsonl_file_path = './eval_output/eval_predictions.jsonl'

# Initialize empty lists to store data
premises = []
hypotheses = []
labels = []
predicted_scores = []
predicted_labels = []

# Read JSONL file line by line
with open(jsonl_file_path, 'r') as file:
    for line in file:
        data = json.loads(line)

        # Append data to lists
        premises.append(data['premise'])
        hypotheses.append(data['hypothesis'])
        labels.append(data['label'])
        predicted_scores.append(data['predicted_scores'])
        predicted_labels.append(data['predicted_label'])

# Create DataFrame
eval_df = pd.DataFrame({
    'premise': premises,
    'hypothesis': hypotheses,
    'label': labels,
    'predicted_scores': predicted_scores,
    'predicted_label': predicted_labels
})
```

```python
eval_df.head()
```

```python
eval_df[eval_df['label'] != eval_df['predicted_label']]
```

```python
def get_tense_from_dataframe(df, premise_col='premise', hypothesis_col='hypothesis'):
    nlp = spacy.load("en_core_web_sm")

    def get_tense(sentence):
        doc = nlp(sentence)
        for token in doc:
            if 'VERB' in token.pos_:
                return token.morph.get('Tense', 'N/A')[0]
        return 'N/A'

    df['premise_tense'] = df[premise_col].apply(get_tense)
    df['hypothesis_tense'] = df[hypothesis_col].apply(get_tense)

    return df
```

```python
eval_df_with_tense = get_tense_from_dataframe(eval_df, premise_col='premise', hypothesis_col='hypothesis')
```

```python
eval_df_with_tense.head()
```

```python
overall_accuracy = (eval_df_with_tense['label'] == eval_df_with_tense['predicted_label']).mean() * 100
print(f"Overall Accuracy: {overall_accuracy:.2f}%")
```

```python
eval_df_with_tense.shape
```

```python
# Filter rows where 'premise_tense' is not equal to 'hypothesis_tense'
filtered_df = eval_df_with_tense[((eval_df_with_tense['premise_tense'] == "Pres") & (\
      eval_df_with_tense['hypothesis_tense'] == "Past")) | \
       ((eval_df_with_tense['premise_tense'] == "Past") & (\
      eval_df_with_tense['hypothesis_tense'] == "Pres"))]

# Calculate accuracy for the filtered DataFrame
filtered_accuracy = (filtered_df['label'] == filtered_df['predicted_label']).mean() * 100
print(f"Accuracy when premise_tense != hypothesis_tense: {filtered_accuracy:.2f}%")
```
